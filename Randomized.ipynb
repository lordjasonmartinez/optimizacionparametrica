```python
import pandas as pd

from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
if __name__ == "__main__":
    dataset = pd.read_csv('./data/diamonds.csv')
    print(dataset)

    X = dataset.drop(['cut', 'color', 'clarity'], axis=1)
    y = dataset['price']  # Cambiado a una sola columna

    reg = RandomForestRegressor()

    parametros = {
        'n_estimators': range(4, 16),
        'criterion': ['friedman_mse', 'squared_error', 'poisson', 'absolute_error'],  # Corregido aqu
        'max_depth': range(2, 11)
    }

    rand_est = RandomizedSearchCV(reg, parametros, n_iter=10, cv=3, scoring='neg_mean_absolute_error').fit(X, y)
    
    print("="*64)
    print(rand_est.best_estimator_)
    print("="*64)
    print(rand_est.best_params_)
    print("="*64)
    print(rand_est.predict(X.loc[[0]]))
    print("="*64)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

if __name__ == "__main__":
    dataset = pd.read_csv('./data/diamonds.csv')
    print(dataset)

    X = dataset.drop(['cut', 'color', 'clarity'], axis=1)
    y = dataset['price']  # Cambiado a una sola columna

    reg = RandomForestRegressor()

    parametros = {
        'n_estimators': range(4, 16),
        'criterion': ['friedman_mse', 'squared_error', 'poisson', 'absolute_error'],  # Corregido aqu
        'max_depth': range(2, 11)
    }

    rand_est = RandomizedSearchCV(reg, parametros, n_iter=10, cv=3, scoring='neg_mean_absolute_error').fit(X, y)

    print("="*64)
    print(rand_est.best_estimator_)
    print("="*64)
    print(rand_est.best_params_)
    print("="*64)
    print(rand_est.predict(X.loc[[0]]))
    print("="*64)

    # Grfico con los resultados
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=range(len(rand_est.cv_results_['mean_test_score'])), y=rand_est.cv_results_['mean_test_score'], color='#FFCE30')
    plt.title('Resultados de la Validacin Cruzada')
    plt.xlabel('Iteracin')
    plt.ylabel('Neg Mean Absolute Error')
    plt.show()
```
